```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tools = ['Tangram', 'CytoSpace', 'cell2location', 'ctopt']
top_100_L5 = [64.31, 67.71, 73.71,76.35]
top_100_whole = [50.25, 62.06, 64.94,66.28]
top_100_visium = [ 54.15, 54.74, 9.74, 70.14]
top_100_kidney = [71.7, 70.34, 65.43, 72.86]


data = pd.DataFrame(columns=['Algorithm', 'Dataset', 'Retention %'])
data.loc[:, 'Algorithm'] = tools * 4
data.loc[:, 'Dataset'] = ['Adult mouse brain'] * len(tools) + ['Mouse embryo whole brain'] * len(tools) + ['Visium mouse brain'] * len(tools) + ['Mouse kidney Slide-seq-V2'] * len(tools)
data.loc[:, 'Retention %'] = top_100_L5 + top_100_whole + top_100_visium + top_100_kidney

data

fig, ax = plt.subplots(figsize=(12,9))
sns.barplot(data=data, x='Dataset', y='Retention %', hue='Algorithm', palette='Accent', ax=ax)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper right')
ax.set_frame_on(False)
# plt.title('Marker genes retention rate', size=16)
plt.savefig('retention.png', bbox_inches='tight', dpi=300)
```
